### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
# Calcular facturaci√≥n total por tienda
facturacion_tienda1 = tienda['Precio'].sum()  # ‚ö†Ô∏è Aqu√≠ es 'tienda', no 'tienda1'
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print("=== FACTURACI√ìN TOTAL POR TIENDA ===")
print(f"Tienda 1: ${facturacion_tienda1:,.0f}")
print(f"Tienda 2: ${facturacion_tienda2:,.0f}")
print(f"Tienda 3: ${facturacion_tienda3:,.0f}")
print(f"Tienda 4: ${facturacion_tienda4:,.0f}")

# Crear gr√°fico
import matplotlib.pyplot as plt
import seaborn as sns

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [facturacion_tienda1, facturacion_tienda2,
           facturacion_tienda3, facturacion_tienda4]

plt.figure(figsize=(10, 6))
sns.barplot(x=tiendas, y=valores, palette='viridis')
plt.title('Facturaci√≥n Total por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Facturaci√≥n ($)', fontsize=12)
plt.xticks(fontsize=10)

# Agregar valores en las barras
for i, v in enumerate(valores):
    plt.text(i, v + 100000, f"${v:,.0f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()

# Identificar tienda con menor facturaci√≥n
menor_facturacion = min(valores)
tienda_menor = tiendas[valores.index(menor_facturacion)]
print(f"\nüî¥ La tienda con MENOR facturaci√≥n es: {tienda_menor}")
```

# 2. Ventas por categor√≠a

```python
# Analizar categor√≠as m√°s populares por tienda
print("=== CATEGOR√çAS M√ÅS POPULARES POR TIENDA ===\n")

def analizar_categorias(df_tienda, nombre_tienda):
    print(f"--- {nombre_tienda} ---")
    categorias = df_tienda.groupby('Categor√≠a del Producto').agg({
        'Precio': 'sum',
        'Producto': 'count'
    }).round(2)

    categorias.columns = ['Facturaci√≥n Total', 'Cantidad Ventas']
    categorias = categorias.sort_values('Facturaci√≥n Total', ascending=False)

    print(categorias.head(3))  # Top 3 categor√≠as
    print("\n")
    return categorias

# Analizar cada tienda
cat_tienda1 = analizar_categorias(tienda, "Tienda 1")
cat_tienda2 = analizar_categorias(tienda2, "Tienda 2")
cat_tienda3 = analizar_categorias(tienda3, "Tienda 3")
cat_tienda4 = analizar_categorias(tienda4, "Tienda 4")

# Gr√°fico comparativo - Top categor√≠a por tienda
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

tiendas_data = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, (df, nombre) in enumerate(zip(tiendas_data, nombres)):
    top_categorias = df.groupby('Categor√≠a del Producto')['Precio'].sum().sort_values(ascending=False).head(3)

    sns.barplot(x=top_categorias.index, y=top_categorias.values,
                ax=axes[i], palette='Set2')
    axes[i].set_title(f'Top 3 Categor√≠as - {nombre}', fontsize=11, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_ylabel('Facturaci√≥n ($)')

plt.tight_layout()
plt.show()
```

# 3. Calificaci√≥n promedio de la tienda


```python
# Calcular calificaci√≥n promedio por tienda
calificacion_tienda1 = tienda['Calificaci√≥n'].mean()
calificacion_tienda2 = tienda2['Calificaci√≥n'].mean()
calificacion_tienda3 = tienda3['Calificaci√≥n'].mean()
calificacion_tienda4 = tienda4['Calificaci√≥n'].mean()

print("=== CALIFICACI√ìN PROMEDIO POR TIENDA ===")
print(f"Tienda 1: {calificacion_tienda1:.2f} / 5.0")
print(f"Tienda 2: {calificacion_tienda2:.2f} / 5.0")
print(f"Tienda 3: {calificacion_tienda3:.2f} / 5.0")
print(f"Tienda 4: {calificacion_tienda4:.2f} / 5.0")

# Gr√°fico de calificaciones
import matplotlib.pyplot as plt
import seaborn as sns

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [calificacion_tienda1, calificacion_tienda2,
           calificacion_tienda3, calificacion_tienda4]

plt.figure(figsize=(10, 6))
bars = sns.barplot(x=tiendas, y=valores, palette='coolwarm')
plt.title('Calificaci√≥n Promedio de Clientes por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificaci√≥n (sobre 5)', fontsize=12)
plt.ylim(0, 5)
plt.axhline(y=4, color='green', linestyle='--', alpha=0.7, label='Excelente (4.0)')
plt.legend()

# Agregar valores en las barras
for bar, val in zip(bars.patches, valores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
             f'{val:.2f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# Tienda con mejor y peor calificaci√≥n
mejor_calif = max(valores)
peor_calif = min(valores)
tienda_mejor = tiendas[valores.index(mejor_calif)]
tienda_peor = tiendas[valores.index(peor_calif)]

print(f"\n‚≠ê Mejor calificaci√≥n: {tienda_mejor} ({mejor_calif:.2f})")
print(f"‚ùå Peor calificaci√≥n: {tienda_peor} ({peor_calif:.2f})")
```

# 4. Productos m√°s y menos vendidos

```python
# Analizar productos m√°s y menos vendidos por tienda
print("=== PRODUCTOS M√ÅS Y MENOS VENDIDOS POR TIENDA ===\n")

def analizar_productos(df_tienda, nombre_tienda):
    print(f"--- {nombre_tienda} ---")

    productos_venta = df_tienda.groupby('Producto').agg({
        'Precio': 'sum',
        'Producto': 'count'
    })
    productos_venta.columns = ['Facturaci√≥n Total', 'Cantidad Vendida']

    # Top 5 m√°s vendidos
    print("\nüî• TOP 5 PRODUCTOS M√ÅS VENDIDOS:")
    print(productos_venta.sort_values('Cantidad Vendida', ascending=False).head(5))

    # Bottom 5 menos vendidos
    print("\n‚ùÑÔ∏è TOP 5 PRODUCTOS MENOS VENDIDOS:")
    print(productos_venta.sort_values('Cantidad Vendida', ascending=True).head(5))

    print("\n")
    return productos_venta

# Analizar cada tienda
prod_tienda1 = analizar_productos(tienda, "Tienda 1")
prod_tienda2 = analizar_productos(tienda2, "Tienda 2")
prod_tienda3 = analizar_productos(tienda3, "Tienda 3")
prod_tienda4 = analizar_productos(tienda4, "Tienda 4")

# Gr√°fico comparativo - Producto m√°s vendido por tienda
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 4, figsize=(16, 4))

tiendas_data = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, (df, nombre) in enumerate(zip(tiendas_data, nombres)):
    producto_mas_vendido = df['Producto'].value_counts().head(1)
    producto_nombre = producto_mas_vendido.index[0]
    producto_valor = producto_mas_vendido.values[0]

    sns.barplot(x=[producto_nombre[:20]], y=[producto_valor],
                ax=axes[i], palette='Set1')
    axes[i].set_title(f'{nombre}\nProducto m√°s vendido', fontsize=10, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_ylabel('Cantidad Vendida')

plt.tight_layout()
plt.show()
```

# 5. Env√≠o promedio por tienda

```python
# Calcular costo promedio de env√≠o por tienda
envio_tienda1 = tienda['Costo de env√≠o'].mean()
envio_tienda2 = tienda2['Costo de env√≠o'].mean()
envio_tienda3 = tienda3['Costo de env√≠o'].mean()
envio_tienda4 = tienda4['Costo de env√≠o'].mean()

print("=== COSTO PROMEDIO DE ENV√çO POR TIENDA ===")
print(f"Tienda 1: ${envio_tienda1:,.0f}")
print(f"Tienda 2: ${envio_tienda2:,.0f}")
print(f"Tienda 3: ${envio_tienda3:,.0f}")
print(f"Tienda 4: ${envio_tienda4:,.0f}")

# Gr√°fico
import matplotlib.pyplot as plt
import seaborn as sns

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]

plt.figure(figsize=(10, 6))
sns.barplot(x=tiendas, y=valores, palette='coolwarm')
plt.title('Costo Promedio de Env√≠o por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo de Env√≠o Promedio ($)', fontsize=12)

# Agregar valores en las barras
for i, v in enumerate(valores):
    plt.text(i, v + 500, f"${v:,.0f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()

# Tienda con mayor y menor costo de env√≠o
mayor_envio = max(valores)
menor_envio = min(valores)
tienda_mayor = tiendas[valores.index(mayor_envio)]
tienda_menor = tiendas[valores.index(menor_envio)]

print(f"\nüì¶ Mayor costo de env√≠o: {tienda_mayor} (${mayor_envio:,.0f})")
print(f"üì¶ Menor costo de env√≠o: {tienda_menor} (${menor_envio:,.0f})")
```

# üéØ CONCLUSI√ìN Y RECOMENDACI√ìN FINAL

```python
print("="*70)
print("üéØ INFORME FINAL PARA EL SR. JUAN - ALURA STORE")
print("="*70)

# Recopilar todos los resultados
resumen = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturaci√≥n': [facturacion_tienda1, facturacion_tienda2,
                    facturacion_tienda3, facturacion_tienda4],
    'Calificaci√≥n': [calificacion_tienda1, calificacion_tienda2,
                     calificacion_tienda3, calificacion_tienda4],
    'Costo_Envio_Promedio': [envio_tienda1, envio_tienda2,
                              envio_tienda3, envio_tienda4]
}

import pandas as pd
df_resumen = pd.DataFrame(resumen)
print("\nüìä RESUMEN GENERAL:")
print(df_resumen.to_string(index=False))

# Calcular puntuaci√≥n de rendimiento
# Menor facturaci√≥n + menor calificaci√≥n = peor desempe√±o
df_resumen['Score_Rendimiento'] = (
    df_resumen['Facturaci√≥n'].rank() +
    df_resumen['Calificaci√≥n'].rank()
)

print("\n" + "="*70)
print("üìà PUNTUACI√ìN DE RENDIMIENTO (mayor = mejor):")
for i, row in df_resumen.iterrows():
    print(f"{row['Tienda']}: {row['Score_Rendimiento']:.0f}")

# Identificar la tienda con peor desempe√±o
tienda_peor = df_resumen.loc[df_resumen['Score_Rendimiento'].idxmin()]

print("\n" + "="*70)
print("‚úÖ RECOMENDACI√ìN FINAL:")
print("="*70)
print(f"\nüî¥ EL SR. JUAN DEBER√çA VENDER: {tienda_peor['Tienda'].upper()}")
print("\nüìã Razones principales:")
print(f"   ‚Ä¢ Facturaci√≥n: ${tienda_peor['Facturaci√≥n']:,.0f}")
print(f"   ‚Ä¢ Calificaci√≥n clientes: {tienda_peor['Calificaci√≥n']:.2f}/5.0")
print(f"   ‚Ä¢ Costo env√≠o promedio: ${tienda_peor['Costo_Envio_Promedio']:,.0f}")
print("\nüí° Esta tienda presenta el menor rendimiento general y ser√≠a")
print("   la mejor candidata para vender e invertir en un nuevo negocio.")
print("="*70)
```
